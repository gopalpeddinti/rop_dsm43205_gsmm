Let's challenge this model for growth on CO2 as the sole carbon source

```python
rho_co2 = rho_cobra.copy()
```

Introducing CO2 exchange to equal carbon moles as used for glucose simulations

```python
rho_co2.reactions.get_by_id('EX_glc__D_e').lower_bound=0
rho_co2.reactions.get_by_id('EX_co2_e').lower_bound=-6
```

Introducing H2 hydrogenase, H2 transport and exchange

```python
reaction = cobra.Reaction('H2HYDROGENASE')
reaction.name = 'H2HYDROGENASE'
reaction.subsystem = ' '
reaction.lower_bound = 0.  # This is the default
reaction.upper_bound = 1000.  # This is the default
```

```python
h2_c=rho_co2.metabolites.get_by_id('h2_c')
nad_c=rho_co2.metabolites.get_by_id('nad_c')
nadh_c=rho_co2.metabolites.get_by_id('nadh_c')
h_c=rho_co2.metabolites.get_by_id('h_c')
```

```python
reaction.add_metabolites({
    h2_c: -1.0,
    nad_c: -1.0,
    nadh_c: 1.0,
    h_c: 1.0
})
rho_co2.add_reactions([reaction])
reaction.reaction
```

```python
reaction = cobra.Reaction('H2t')
reaction.name = 'H2t'
reaction.subsystem = ' '
reaction.lower_bound = -1000.
reaction.upper_bound = 1000.  # This is the default
```

```python
h_e=rho_co2.metabolites.get_by_id('h_e')
```

```python
h2_e = cobra.Metabolite(
    'h2_e',
    formula='H2',
    name='Hydrogen',
    compartment='e')
```

```python
reaction.add_metabolites({
    h2_e: -1.0,
    h2_c: 1.0
})
rho_co2.add_reactions([reaction])
reaction.reaction
```

```python
reaction = cobra.Reaction('EX_h2_e')
reaction.name = 'EX_h2_e'
reaction.subsystem = ' '
reaction.lower_bound = -1000.
reaction.upper_bound = 1000.  # This is the default
```

```python
reaction.add_metabolites({
    h2_e: -1.0,
})
rho_co2.add_reactions([reaction])
reaction.reaction
```

Introducing reactions for lactic acid production.

```python
reaction = cobra.Reaction('lac__L_transport')
reaction.name = 'L-lactate_transport'
reaction.subsystem = ' '
reaction.lower_bound = -1000.
reaction.upper_bound = 1000.  # This is the default
```

```python
lac__L_e = cobra.Metabolite(
    'lac__L_e',
    formula='C3H6O3',
    name='L-lactate',
    compartment='e')
```

```python
lac__L_c=rho_co2.metabolites.get_by_id('lac__L_c')
```

```python
reaction.add_metabolites({
    lac__L_c: -1.0,
    lac__L_e: 1.0,
})
rho_co2.add_reactions([reaction])
reaction.reaction
```

```python
reaction = cobra.Reaction('EX_lac__L')
reaction.name = 'L-lactate_exchange'
reaction.subsystem = ' '
reaction.lower_bound = 0.
reaction.upper_bound = 1000.  # This is the default
```

```python
reaction.add_metabolites({
    lac__L_e: -1.0
})
rho_co2.add_reactions([reaction])
reaction.reaction
```

```python
rho_co2.objective="EX_lac__L"
```

```python
solution_co2=rho_co2.optimize()
print('Objective:', solution_co2['EX_lac__L'])
```

```python
rho_co2.metabolites.pyr_c.summary()
```

```python
rho_co2.reactions.get_by_id('EX_lac__L').upper_bound = numpy.floor(numpy.power(10,10)*solution_co2['EX_lac__L'])/numpy.power(10,10)
rho_co2.reactions.get_by_id('EX_lac__L').lower_bound = numpy.floor(numpy.power(10,10)*solution_co2['EX_lac__L'])/numpy.power(10,10)
minmax_co2_lac=flux_variability_analysis(rho_co2,fraction_of_optimum=1)
```

```python
print(minmax_co2_lac)
```

```python
minmax_co2_lac.loc['EX_co2_e']
```

Calculating the H2 to CO2 utilization ratio for optimal lactic acid synthesis

```python
h2_co2_ratio_lac = 10/minmax_co2_lac.loc['EX_co2_e'][1]
print(h2_co2_ratio_lac)
```
